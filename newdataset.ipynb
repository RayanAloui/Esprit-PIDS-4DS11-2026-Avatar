```python
import requests
import pandas as pd
import time

cities = {
    "Tunis": (36.8065, 10.1815),
}

data = []

# 🔥 plusieurs serveurs (fallback)
urls = [
    "https://overpass-api.de/api/interpreter",
    "https://lz4.overpass-api.de/api/interpreter",
    "https://overpass.kumi.systems/api/interpreter"
]

for city, (lat, lon) in cities.items():
    print(f"Processing {city}...")

    query = f"""
[out:json][timeout:25];
node["amenity"="pharmacy"](around:4000,{lat},{lon});
out;
"""

    success = False

    for url in urls:  # 🔥 essayer plusieurs serveurs
        try:
            response = requests.get(url, params={'data': query}, timeout=30)

            if response.status_code == 200 and response.text.strip():
                result = response.json()
                pharmacies = result.get('elements', [])

                count = 0

                for p in pharmacies:
                    if 'tags' in p and 'name' in p['tags']:
                        data.append({
                            "city": city,
                            "name": p['tags']['name'],
                            "latitude": p['lat'],
                            "longitude": p['lon']
                        })
                        count += 1

                    if count >= 25:
                        break

                success = True
                break

        except:
            continue

    if not success:
        print(f"❌ Impossible pour {city}")

    time.sleep(3)  # 🔥 très important

df = pd.DataFrame(data)
df = df.drop_duplicates(subset=["name", "latitude", "longitude"])

print(df)
import os

file_name = "Pharmaciess.csv"

if os.path.exists(file_name):
    # 🔥 ajouter sans écraser
    df.to_csv(file_name, mode='a', index=False, header=False, encoding="utf-8-sig",sep=";")
else:
    # 🔥 créer fichier avec header
    df.to_csv(file_name, index=False, encoding="utf-8-sig",sep=";")

print("✅ Données ajoutées au fichier")
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler

# ===============================
# 1. LOAD DATASETS
# ===============================
df_old = pd.read_csv("pharmacies_foot_traffic.csv")
df_new = pd.read_csv("Pharmaciess.csv", encoding="utf-8-sig", sep=";")

# ===============================
# 2. CLEAN DATA
# ===============================

# 🔹 df_new
df_new = df_new.drop_duplicates(subset=["name", "latitude", "longitude"])
df_new = df_new.dropna(subset=["latitude", "longitude"])
df_new = df_new.reset_index(drop=True)

# 🔹 df_old
df_old["day_text"] = df_old["day_text"].astype(str).str.strip().str.lower()
df_old = df_old[df_old["day_text"] != "nan"]
df_old = df_old.dropna(subset=["day_text", "latitude", "longitude"])

print("✅ Cleaning done")

# ===============================
# 3. FEATURES
# ===============================
features = ["latitude", "longitude"]

targets = [
    'hour_0','hour_1','hour_2','hour_3','hour_4','hour_5',
    'hour_6','hour_7','hour_8','hour_9','hour_10','hour_11',
    'hour_12','hour_13','hour_14','hour_15','hour_16','hour_17',
    'hour_18','hour_19','hour_20','hour_21','hour_22','hour_23',
    'day_mean','day_max'
]

# ===============================
# 4. KNN PROCESS
# ===============================
scaler = StandardScaler()
df_result = []

unique_days = df_old["day_text"].unique()

for day in unique_days:

    print(f"Processing {day}")

    df_day = df_old[df_old["day_text"] == day]

    if df_day.shape[0] == 0:
        continue

    X_new = scaler.fit_transform(df_new[features])

    temp_df = df_new.copy()
    temp_df["day_text"] = day

    for target in targets:

        if target not in df_day.columns:
            continue

        # 🔥 supprimer NaN
        df_temp = df_day.dropna(subset=[target])

        if df_temp.shape[0] < 5:
            continue

        X_train = scaler.fit_transform(df_temp[features])
        y_train = df_temp[target]

        knn = KNeighborsRegressor(n_neighbors=5, weights="distance")
        knn.fit(X_train, y_train)

        temp_df[target] = knn.predict(X_new)

    df_result.append(temp_df)

# ===============================
# 5. CONCAT FINAL
# ===============================
df_final = pd.concat(df_result, ignore_index=True)

print("✅ KNN terminé")

# ===============================
# 6. AJOUT LOGIQUE MÉTIER 🔥
# ===============================

# 🔹 détecter activité nocturne
night_hours = ['hour_0','hour_1','hour_2','hour_3','hour_4','hour_5']

df_final["night_activity"] = df_final[night_hours].mean(axis=1)

# 🔹 classifier pharmacie
df_final["type"] = df_final["night_activity"].apply(
    lambda x: "night" if x > 20 else "day"
)

# 🔹 corriger pharmacies de jour
for i in range(24):
    col = f"hour_{i}"

    if i < 8 or i > 20:
        df_final.loc[df_final["type"] == "day", col] = 0

# 🔹 nettoyage final
df_final = df_final.fillna(0)

# ===============================
# 7. SAVE
# ===============================
df_final.to_excel("dataset_final_knn_clean.xlsx", index=False)

print("✅ DATASET FINAL CLEAN CRÉÉ")
```

```python
import pandas as pd

# ===============================
# 1. LOAD DATA
# ===============================
df_final = pd.read_excel("dataset_final_knn_clean.xlsx")
df_extra = pd.read_csv("pharmacy_doctor_database.csv")

# ===============================
# 2. CLEAN NAMES (IMPORTANT)
# ===============================
df_final["name"] = df_final["name"].str.strip().str.lower()
df_extra["client_name"] = df_extra["client_name"].str.strip().str.lower()

# ===============================
# 3. SELECT FEATURES
# ===============================
df_extra_selected = df_extra[[
    "client_name",
    "patients_per_month",
    "last_visit_date",
    "total_visits",
    "engagement_level"
]]

# ===============================
# 4. CONVERT DATE
# ===============================
df_extra_selected["last_visit_date"] = pd.to_datetime(
    df_extra_selected["last_visit_date"],
    errors="coerce"
)

# ===============================
# 5. MERGE
# ===============================
df_final = df_final.merge(
    df_extra_selected,
    left_on="name",
    right_on="client_name",
    how="left"
)

# ===============================
# 6. FEATURE ENGINEERING 🔥
# ===============================

# days since last visit
today = pd.Timestamp.today()

df_final["days_since_last_visit"] = (
    today - df_final["last_visit_date"]
).dt.days

# ===============================
# 7. CLEAN MISSING VALUES
# ===============================
df_final["patients_per_month"] = df_final["patients_per_month"].fillna(0)
df_final["total_visits"] = df_final["total_visits"].fillna(0)
df_final["days_since_last_visit"] = df_final["days_since_last_visit"].fillna(999)

# ===============================
# 8. SAVE
# ===============================
df_final.to_excel("dataset_final_enriched.xlsx", index=False)

print("✅ DATASET ENRICHI CRÉÉ")
```

```python
import pandas as pd
import numpy as np

# ===============================
# 1. LOAD DATASET
# ===============================
df = pd.read_excel("dataset_final_knn_clean.xlsx")

# nettoyer colonnes
df.columns = df.columns.str.strip().str.lower()

# ===============================
# 2. CREATE TOTAL VISITS 🔥
# ===============================

np.random.seed(42)

if "total_visits" not in df.columns:
    df["total_visits"] = np.random.randint(1, 15, len(df))
else:
    df["total_visits"] = np.where(
        df["total_visits"] == 0,
        np.random.randint(1, 15, len(df)),
        df["total_visits"]
    )

# ===============================
# 3. CREATE DAYS SINCE LAST VISIT 🔥
# ===============================

if "days_since_last_visit" not in df.columns:
    df["days_since_last_visit"] = np.random.randint(1, 60, len(df))
else:
    df["days_since_last_visit"] = np.where(
        df["days_since_last_visit"] == 999,
        np.random.randint(1, 60, len(df)),
        df["days_since_last_visit"]
    )

# ===============================
# 4. CREATE LAST VISIT DATE 🔥
# ===============================

today = pd.Timestamp.today()

df["last_visit_date"] = today - pd.to_timedelta(
    df["days_since_last_visit"], unit="d"
)

# ===============================
# 5. AJOUT COHÉRENCE MÉTIER 🔥
# ===============================

# patients
df["patients_per_month"] = np.random.randint(50, 500, len(df))

# logique DSO : + patients → visite récente
df["days_since_last_visit"] = (
    60 - (df["patients_per_month"] / df["patients_per_month"].max() * 50)
).astype(int)

# recalcul date
df["last_visit_date"] = today - pd.to_timedelta(
    df["days_since_last_visit"], unit="d"
)

# ===============================
# 6. CLEAN FINAL
# ===============================
df = df.fillna(0)

# ===============================
# 7. SAVE
# ===============================
df.to_excel("dataset_fixed_features.xlsx", index=False)

print("🔥 FEATURES AJOUTÉES SANS ERREUR")
```
