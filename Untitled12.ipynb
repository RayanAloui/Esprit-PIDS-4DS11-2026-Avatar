```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from geopy.distance import geodesic

# ===============================
# 1. LOAD DATA
# ===============================
df = pd.read_excel("dataset_fixed_features.xlsx")
df.columns = df.columns.str.strip().str.lower()

# ===============================
# 2. PARAMÈTRES
# ===============================
day = "monday"
hour = "hour_10"
user_location = (36.8065, 10.1815)

# ===============================
# 3. FILTER JOUR
# ===============================
df_day = df[df["day_text"] == day].copy()

# ===============================
# 4. TRAFFIC + FILTRAGE 🔥
# ===============================
df_day["traffic"] = df_day[hour]

# ❌ supprimer pharmacies fermées
df_day = df_day[df_day["traffic"] > 0]

# ===============================
# 5. DISTANCE
# ===============================
def compute_distance(row):
    return geodesic(user_location, (row["latitude"], row["longitude"])).km

df_day["distance"] = df_day.apply(compute_distance, axis=1)

# ===============================
# 6. NORMALISATION
# ===============================
scaler = MinMaxScaler()

features = [
    "distance",
    "traffic",
    "patients_per_month",
    "days_since_last_visit"
]

df_day[[f + "_norm" for f in features]] = scaler.fit_transform(df_day[features])

# ===============================
# 7. TRANSFORMATION (CLÉ 🔥)
# ===============================

# inverser distance et traffic
df_day["distance_score"] = 1 - df_day["distance_norm"]
df_day["traffic_score"] = 1 - df_day["traffic_norm"]

# ===============================
# 8. SCORE FINAL (ADDITIF 🔥🔥)
# ===============================

w_distance = 0.3
w_traffic = 0.4
w_patients = 0.2
w_lastvisit = 0.1

df_day["score"] = (
    w_distance * df_day["distance_score"] +
    w_traffic * df_day["traffic_score"] +
    w_patients * df_day["patients_per_month_norm"] +
    w_lastvisit * df_day["days_since_last_visit_norm"]
)

print("🔥 Score calculé (positif et additif)")

# ===============================
# 9. TOP PHARMACIES
# ===============================
top_pharmacies = df_day.sort_values("score", ascending=False).head(10)

print(top_pharmacies[[
    "name",
    "distance",
    "traffic",
    "patients_per_month",
    "days_since_last_visit",
    "score"
]])

# ===============================
# 10. SAVE
# ===============================
top_pharmacies.to_excel(f"top_{day}_{hour}.xlsx", index=False)

print("✅ TOP pharmacies sauvegardées")
```

```python
import pandas as pd
import folium
from geopy.distance import geodesic

# ===============================
# 1. LOAD TOP PHARMACIES
# ===============================
df = pd.read_excel("top_monday_hour_10.xlsx")

# ===============================
# 2. POINT DE DÉPART
# ===============================
start_location = (36.8065, 10.1815)

# ===============================
# 3. ROUTE OPTIMALE (GREEDY 🔥)
# ===============================
remaining = df.copy()
current = start_location
route = []

while len(remaining) > 0:
    remaining["dist"] = remaining.apply(
        lambda row: geodesic(current, (row["latitude"], row["longitude"])).km,
        axis=1
    )

    next_point = remaining.loc[remaining["dist"].idxmin()]
    
    route.append(next_point)
    
    current = (next_point["latitude"], next_point["longitude"])
    
    remaining = remaining.drop(next_point.name)

route_df = pd.DataFrame(route).reset_index(drop=True)

print("🔥 Route optimisée créée")

# ===============================
# 4. MAP INTERACTIVE 🔥
# ===============================
map_route = folium.Map(location=start_location, zoom_start=11)

# départ
folium.Marker(
    location=start_location,
    popup="🏁 Départ",
    icon=folium.Icon(color="green")
).add_to(map_route)

coordinates = [start_location]

# ===============================
# 5. AJOUT POINTS AVEC SCORE 🔥
# ===============================
for i, row in route_df.iterrows():
    loc = (row["latitude"], row["longitude"])
    coordinates.append(loc)

    folium.Marker(
        location=loc,
        popup=f"""
        <b>{i+1}. {row['name']}</b><br>
        Score: {row['score']:.2f}<br>
        Traffic: {row['traffic']}<br>
        Distance: {row['distance']:.2f} km
        """,
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(map_route)

# ===============================
# 6. TRAJET 🔥
# ===============================
folium.PolyLine(
    coordinates,
    color="red",
    weight=4,
    opacity=0.8
).add_to(map_route)

# ===============================
# 7. SAVE
# ===============================
map_route.save("route_optimale_finale.html")

print("🚀 MAP FINALE CRÉÉE : route_optimale_finale.html")
```
