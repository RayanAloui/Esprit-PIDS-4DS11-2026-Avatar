## 1. Load Data & Setup

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
from pathlib import Path
from collections import Counter

# Paths
dataset_path = Path('gesture_dataset')
annotations_path = dataset_path / 'annotations.csv'
metadata_path = dataset_path / 'metadata.json'

# Load annotations
df = pd.read_csv(annotations_path)

# Load metadata
with open(metadata_path, 'r', encoding='utf-8') as f:
    metadata = json.load(f)

print(f"Dataset loaded: {len(df)} images, {len(df.columns)} columns")
print(f"Dataset name: {metadata['dataset_name']}")
````

````python
# Display metadata
print("\nDataset Metadata:")
print("="*60)
print(f"Name: {metadata['dataset_name']}")
print(f"Purpose: {metadata['purpose']}")
print(f"Total images: {metadata['total_images']}")
print(f"Image format: {metadata['image_format']}")
print(f"Image size: {metadata['image_size']}")
print(f"Use case: {metadata['use_case']}")
````

### Column Descriptions:

| Column | Description |
|--------|-------------|
| **image_id** | Unique image identifier |
| **filename** | Image filename |
| **category** | Gesture category (confident, hesitant, etc.) |
| **description** | Description of gesture type |
| **primary_label** | Main label for the gesture |
| **secondary_labels** | Additional related labels |
| **confidence_score** | Annotation confidence (0-1) |
| **annotator** | Person who annotated the image |
| **width** | Image width in pixels |
| **height** | Image height in pixels |
| **format** | Image file format |

## 3. Check Data Quality

````python
# Check missing values
print("Missing Values:")
print(df.isnull().sum())
print(f"\nMissing percentage:")
print((df.isnull().sum() / len(df) * 100).round(1))
````

````python
# Check duplicates
print(f"Duplicate rows: {df.duplicated().sum()}")
print(f"Duplicate image IDs: {df['image_id'].duplicated().sum()}")
print(f"Duplicate filenames: {df['filename'].duplicated().sum()}")
````

````python
# Check image dimensions consistency
print("\nImage Dimensions:")
print(f"Unique widths: {df['width'].unique()}")
print(f"Unique heights: {df['height'].unique()}")
print(f"→ All images have consistent dimensions: {len(df['width'].unique()) == 1 and len(df['height'].unique()) == 1}")
````

---
## 4. Analyze Gesture Categories

````python
# Category distribution
print("Gesture Categories:")
print(df['category'].value_counts())
print(f"\nTotal categories: {df['category'].nunique()}")
print(f"\nPercentages:")
print((df['category'].value_counts() / len(df) * 100).round(1))
````

````python
# Visualize category distribution
plt.figure(figsize=(12, 6))
category_counts = df['category'].value_counts()
colors = ['green', 'red', 'blue', 'orange', 'purple', 'pink', 'brown', 'cyan']
category_counts.plot(kind='bar', color=colors)
plt.title('Distribution of Gesture Categories', fontsize=14, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of Images', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

````python
# Category descriptions
print("\nCategory Descriptions:")
for category in df['category'].unique():
    desc = df[df['category'] == category]['description'].iloc[0]
    count = (df['category'] == category).sum()
    print(f"\n• {category.upper()} ({count} images)")
    print(f"  {desc}")
````

---
## 5. Analyze Labels

````python
# Primary labels distribution
print("Primary Labels:")
print(df['primary_label'].value_counts())
print(f"\nUnique primary labels: {df['primary_label'].nunique()}")
````

````python
# Extract all secondary labels
all_secondary_labels = []
for labels in df['secondary_labels'].dropna():
    for label in labels.split(', '):
        all_secondary_labels.append(label.strip())

label_counts = Counter(all_secondary_labels)

print("\nSecondary Labels Distribution:")
for label, count in label_counts.most_common():
    print(f"  {label}: {count}")
````

````python
# Visualize all labels
plt.figure(figsize=(12, 6))
label_df = pd.DataFrame.from_dict(label_counts, orient='index', columns=['Count'])
label_df.sort_values('Count', ascending=True).plot(kind='barh', color='teal', legend=False)
plt.title('Secondary Labels Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Number of Images', fontsize=12)
plt.ylabel('Label', fontsize=12)
plt.tight_layout()
plt.show()
````

---
## 6. Analyze Annotation Quality

````python
# Confidence score statistics
print("Annotation Confidence Scores:")
print(df['confidence_score'].describe())
````

````python
# Visualize confidence distribution
plt.figure(figsize=(10, 6))
df['confidence_score'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Annotation Confidence Scores', fontsize=14, fontweight='bold')
plt.xlabel('Confidence Score', fontsize=12)
plt.ylabel('Number of Images', fontsize=12)
plt.axvline(df['confidence_score'].mean(), color='red', linestyle='--', linewidth=2, 
            label=f'Mean: {df["confidence_score"].mean():.2f}')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Confidence by category
print("\nAverage Confidence by Category:")
conf_by_category = df.groupby('category')['confidence_score'].mean().sort_values(ascending=False)
print(conf_by_category.round(3))
````

````python
# Low confidence images (might need re-annotation)
low_confidence = df[df['confidence_score'] < 0.80]
print(f"\n⚠️ Images with low confidence (<0.80): {len(low_confidence)} ({len(low_confidence)/len(df)*100:.1f}%)")
if len(low_confidence) > 0:
    print("\nCategories with low confidence:")
    print(low_confidence['category'].value_counts())
````

## 7. Analyze Annotators

````python
# Annotator distribution
print("Annotators:")
print(df['annotator'].value_counts())
print(f"\nTotal annotators: {df['annotator'].nunique()}")
````

````python
# Images per annotator
plt.figure(figsize=(10, 6))
df['annotator'].value_counts().plot(kind='bar', color='coral')
plt.title('Images Annotated per Annotator', fontsize=14, fontweight='bold')
plt.xlabel('Annotator', fontsize=12)
plt.ylabel('Number of Images', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

````python
# Inter-annotator agreement (check if same categories have similar confidence)
print("\nAverage Confidence by Annotator:")
conf_by_annotator = df.groupby('annotator')['confidence_score'].agg(['mean', 'std']).round(3)
print(conf_by_annotator)
````

---
## 8. Class Balance Analysis

````python
# Check class balance
print("Class Balance Analysis:")
print("="*60)

category_counts = df['category'].value_counts()
min_class = category_counts.min()
max_class = category_counts.max()
balance_ratio = max_class / min_class

print(f"Smallest class: {category_counts.idxmin()} with {min_class} images")
print(f"Largest class: {category_counts.idxmax()} with {max_class} images")
print(f"Balance ratio: {balance_ratio:.2f}")

if balance_ratio < 1.5:
    print("\n✓ Classes are well balanced (ratio < 1.5)")
elif balance_ratio < 2.0:
    print("\n⚠️ Minor class imbalance (ratio < 2.0)")
else:
    print("\n❌ Significant class imbalance (ratio ≥ 2.0) - Consider data augmentation")
````

````python
# Visualize class balance
plt.figure(figsize=(10, 6))
category_pct = (category_counts / len(df) * 100).sort_values()
category_pct.plot(kind='barh', color='lightgreen')
plt.axvline(x=12.5, color='red', linestyle='--', label='Perfect balance (12.5%)')
plt.title('Class Balance - Percentage of Images per Category', fontsize=14, fontweight='bold')
plt.xlabel('Percentage (%)', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
````

## 9. Relevance to DSO 1 (Training & Evaluation)

### 🎯 How This Dataset Supports DSO 1:

**DSO 1 Objective:** Développer des modèles ML/NLP pour évaluer automatiquement les interactions des délégués médicaux

**Enhancement with Gestures:**
L'évaluation devient **multimodale** :
- ✅ **Verbal** : Contenu, précision scientifique (Dataset #4 - Conversations)
- ✅ **Scores** : Performance historique (Dataset #2 - Performance)
- ✅ **Non-verbal** : Langage corporel, gestes, confiance (Dataset #6 - Gestures) ← **NOUVEAU**

**Use Cases Concrets:**

1. **Détection de confiance/stress**
   - Modèle détecte gestes hésitants → Score confidence ajusté
   - Gestes défensifs → Indicateur de stress/inconfort
   - Gestes enthousiastes → Bonus engagement

2. **Feedback enrichi**
   - "Votre argumentation est solide (8/10) mais vos gestes montrent de l'hésitation"
   - "Excellente présentation ! Vos gestes renforcent votre message"

3. **Adaptive Learning**
   - Si gestes stressés détectés → Réduire difficulté temporairement
   - Si gestes confiants → Augmenter complexité du scénario

4. **Training ciblé**
   - Identifier délégués avec bon verbal mais faible non-verbal
   - Créer exercices spécifiques sur communication corporelle

````python
# Simuler des scores combinés (verbal + non-verbal)
print("SIMULATION - Impact des Gestes sur l'Évaluation Globale:")
print("="*60)

# Définir des mappings geste → impact score
gesture_impact = {
    'confident': +0.5,
    'enthusiastic': +0.4,
    'explaining': +0.3,
    'listening': +0.2,
    'thinking': 0.0,
    'questioning': 0.0,
    'hesitant': -0.3,
    'defensive': -0.5
}

# Calculer impact moyen par catégorie
print("\nImpact des Gestes sur le Score Final:")
for category, impact in gesture_impact.items():
    count = (df['category'] == category).sum()
    symbol = '↑' if impact > 0 else '↓' if impact < 0 else '→'
    print(f"  {symbol} {category.capitalize():<15} : {impact:+.1f} points ({count} images)")

print("\nExemple de Scoring Combiné:")
print("-" * 60)
print("Scénario 1:")
print("  - Score verbal: 7.5/10")
print("  - Gestes détectés: 'confident' → +0.5")
print("  - Score final: 8.0/10 ✓")
print("\nScénario 2:")
print("  - Score verbal: 7.5/10")
print("  - Gestes détectés: 'hesitant' → -0.3")
print("  - Score final: 7.2/10")
print("  - Feedback: 'Votre contenu est bon mais travaillez votre assurance'")
````

## 10. Computer Vision Model Recommendations

### 📊 Recommandations pour la Phase Modeling:

**1. Architectures Recommandées:**
- **Transfer Learning avec CNN pré-entraînés:**
  - ResNet50 / EfficientNet (meilleur choix)
  - MobileNetV2 (si besoin de rapidité)
  - VGG16 (baseline simple)

- **Vision Transformers (avancé):**
  - ViT (Vision Transformer)
  - Swin Transformer

**2. Data Augmentation (CRUCIAL avec 335 images):**
```python
# Augmentation recommandée
- Rotation: ±15°
- Flip horizontal: Oui
- Brightness: ±20%
- Zoom: 0.8-1.2
- Shift: ±10%
```
→ Objectif: Augmenter dataset à ~2000-3000 images

**3. Train/Val/Test Split:**
- Training: 70% (~235 images, + augmentation)
- Validation: 15% (~50 images)
- Test: 15% (~50 images)
- **Important:** Stratifier par catégorie

**4. Metrics d'Évaluation:**
- Accuracy (objectif: >85%)
- F1-Score par classe (objectif: >0.80)
- Confusion Matrix (identifier classes difficiles)
- Top-2 Accuracy (pour gestes ambigus)

**5. Pipeline de Production:**
```
Webcam/Vidéo → Frame Extraction → Preprocessing → CNN → 
Prédiction Geste → Calcul Impact → Score Ajusté → Feedback
```

**6. Optimisations:**
- Utiliser GPU pour training (Google Colab gratuit)
- Batch size: 32 (optimal pour transfer learning)
- Learning rate: 1e-4 avec decay
- Early stopping sur validation loss
- Sauvegarder meilleur modèle (checkpoint)
