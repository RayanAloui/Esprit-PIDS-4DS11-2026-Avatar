## 4. Data Quality Assessment

```python
# Missing values analysis
print("Missing Values Analysis:")
print("="*60)
missing_data = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isnull().sum(),
    'Missing_Percentage': (df.isnull().sum() / len(df) * 100).round(2)
})
missing_data = missing_data[missing_data['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)
missing_data
```

```python
# Visualize missing data
plt.figure(figsize=(10, 6))
missing_pct = (df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
missing_pct[missing_pct > 0].plot(kind='bar', color='coral')
plt.title('Missing Data by Column (%)', fontsize=14, fontweight='bold')
plt.xlabel('Column', fontsize=12)
plt.ylabel('Missing Percentage (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Empty strings check (common in text fields)
print("\nEmpty String Check:")
print("="*60)
for col in df.columns:
    empty_count = (df[col].astype(str).str.strip() == '').sum()
    if empty_count > 0:
        print(f"• {col}: {empty_count} empty strings ({empty_count/len(df)*100:.1f}%)")
```

```python
# Check for duplicates
print("\nDuplicate Analysis:")
print("="*60)
duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {duplicates}")

# Check duplicate product names
duplicate_names = df['name'].duplicated().sum()
print(f"Duplicate product names: {duplicate_names}")

if duplicate_names > 0:
    print("\nDuplicated product names:")
    print(df[df['name'].duplicated(keep=False)][['name', 'categories', 'forme']].sort_values('name'))
```

---
## 5. Descriptive Statistics

```python
# Text length statistics for key columns
print("Text Length Statistics:")
print("="*60)

text_cols = ['name', 'indications', 'categories']
text_stats = pd.DataFrame()

for col in text_cols:
    text_stats[col] = [
        df[col].astype(str).str.len().min(),
        df[col].astype(str).str.len().max(),
        df[col].astype(str).str.len().mean().round(1),
        df[col].astype(str).str.len().median()
    ]

text_stats.index = ['Min Length', 'Max Length', 'Mean Length', 'Median Length']
text_stats
```

```python
# Product form (forme) distribution
print("\nProduct Forms Distribution:")
print("="*60)
forme_dist = df['forme'].value_counts()
print(forme_dist)
print(f"\nTotal unique forms: {df['forme'].nunique()}")
```

```python
# Visualize product forms
plt.figure(figsize=(12, 6))
forme_dist.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Product Forms', fontsize=14, fontweight='bold')
plt.xlabel('Form', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Product class (classe) distribution
print("\nProduct Class Distribution:")
print("="*60)
classe_dist = df['classe'].value_counts()
print(classe_dist)
print(f"\nTotal unique classes: {df['classe'].nunique()}")
```

```python
# Visualize product classes
plt.figure(figsize=(10, 6))
classe_dist.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Product Class Distribution', fontsize=14, fontweight='bold')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

---
## 6. Categories Analysis

```python
# Categories can contain multiple values separated by commas
# Let's analyze them

print("Categories Analysis:")
print("="*60)

# Split categories and count
all_categories = []
for cat in df['categories'].dropna():
    if isinstance(cat, str):
        # Split by comma and clean
        cats = [c.strip() for c in cat.split(',')]
        all_categories.extend(cats)

category_counts = Counter(all_categories)
category_df = pd.DataFrame.from_dict(category_counts, orient='index', columns=['Count'])
category_df = category_df.sort_values('Count', ascending=False)

print(f"Total unique categories: {len(category_df)}")
print(f"\nTop 15 categories:")
category_df.head(15)
```

```python
# Visualize top categories
plt.figure(figsize=(12, 8))
category_df.head(15).plot(kind='barh', color='mediumseagreen', legend=False)
plt.title('Top 15 Health Categories', fontsize=14, fontweight='bold')
plt.xlabel('Number of Products', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Number of categories per product
df['num_categories'] = df['categories'].apply(
    lambda x: len([c.strip() for c in str(x).split(',')]) if pd.notna(x) else 0
)

print("\nNumber of categories per product:")
print(df['num_categories'].value_counts().sort_index())

print(f"\nAverage categories per product: {df['num_categories'].mean():.2f}")
```

---
## 7. Indications Analysis

```python
# Analyze indications text
print("Indications Analysis:")
print("="*60)

# Products with no indications
no_indications = df['indications'].isna().sum()
print(f"Products without indications: {no_indications} ({no_indications/len(df)*100:.1f}%)")

# Products with 'Aucune' (None)
aucune_count = (df['indications'].astype(str).str.strip() == 'Aucune').sum()
print(f"Products with 'Aucune' (None) indications: {aucune_count}")

# Sample of products with detailed indications
print("\nSample products with indications:")
df[df['indications'].notna() & (df['indications'].astype(str).str.len() > 20)][['name', 'indications']].head(3)
```

```python
# Count indication points (lines starting with –)
df['num_indications'] = df['indications'].apply(
    lambda x: str(x).count('–') if pd.notna(x) else 0
)

print("\nNumber of indication points per product:")
print(df['num_indications'].describe())

# Distribution visualization
plt.figure(figsize=(10, 6))
df['num_indications'].value_counts().sort_index().plot(kind='bar', color='lightcoral')
plt.title('Distribution of Indication Points per Product', fontsize=14, fontweight='bold')
plt.xlabel('Number of Indication Points', fontsize=12)
plt.ylabel('Number of Products', fontsize=12)
plt.tight_layout()
plt.show()
```

## 8. Data Quality Summary

```python
print("DATA QUALITY SUMMARY")
print("="*60)

# Completeness
print("\n1. COMPLETENESS:")
print("-" * 40)
for col in df.columns:
    completeness = (1 - df[col].isna().sum() / len(df)) * 100
    status = "Good" if completeness > 80 else "Needs attention"
    print(f"{col:<20}: {completeness:>6.1f}% {status}")

# Uniqueness
print("\n2. UNIQUENESS:")
print("-" * 40)
print(f"Unique products (by name): {df['name'].nunique()} / {len(df)}")
print(f"Duplicate names: {df['name'].duplicated().sum()}")

# Consistency
print("\n3. CONSISTENCY:")
print("-" * 40)
print(f"Unique product forms: {df['forme'].nunique()}")
print(f"Unique product classes: {df['classe'].nunique()}")
print(f"Unique health categories: {len(category_df)}")

# Richness
print("\n4. DATA RICHNESS:")
print("-" * 40)
print(f"Avg categories per product: {df['num_categories'].mean():.2f}")
print(f"Avg indication points per product: {df['num_indications'].mean():.2f}")
print(f"Avg indication text length: {df['indications'].astype(str).str.len().mean():.0f} characters")
```

## 9. Key Insights & Observations

### Key Findings:

**1. Dataset Overview:**
- Total of 215 pharmaceutical products
- 8 columns covering product identification, categorization, and medical information
- Data is primarily textual (product names, indications, categories)

**2. Data Quality:**
- **Missing Data**: Several columns have missing values, particularly `infos_produit` and `classe`
- **Consistency**: Product forms and classes show some variability in naming
- **Richness**: Products have multiple categories and detailed indications

**3. Product Composition:**
- Most products are "Complément alimentaire" (dietary supplements)
- Common forms: Gélules (capsules), Sirop (syrup), Crème (cream)
- Products target diverse health categories (skin, immunity, digestion, etc.)

**4. Medical Information:**
- Rich indication data for most products (multiple benefit points)
- Categories span various medical conditions and health areas
- Some products have "Aucune" (None) or minimal indication data

### Relevance to Avatar Project:

**For Training Mode (DSO 1):**
- Product categories and indications provide realistic scenarios for medical rep training
- Diversity of products allows for varied simulation exercises
- Need to handle products with limited information in training scenarios

**For Commercial Mode (DSO 2):**
- Structured product data enables automated, consistent presentations
- Multi-category products require smart content adaptation
- Rich indication data supports detailed, professional explanations

**For CRM & Analytics (DSO 3):**
- Product categories enable segmentation and targeting analysis
- Form types can help identify prescription patterns
- Medical indications support efficacy tracking

**For Route Optimization (DSO 4):**
- Product portfolio understanding helps prioritize pharmacy visits
- Category data can inform which pharmacies to target based on specialization
