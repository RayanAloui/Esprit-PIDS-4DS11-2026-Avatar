## 1. Setup & Data Loading

````python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

# Configuration
dataset_path = Path('gesture_dataset_kaggle')  # Modifier selon votre chemin

# Vérifier l'existence du dataset
if not dataset_path.exists():
    print("❌ Dataset not found!")
    print(f"Expected location: {dataset_path}")
    print("\nPlease download the dataset from Kaggle:")
    print("https://www.kaggle.com/datasets/sarvaninataraj/gesture-recognition")
    print("\nSee KAGGLE_DOWNLOAD_GUIDE.md for instructions.")
else:
    print("✓ Dataset folder found!")
````

````python
# Définir les catégories de gestes
gesture_categories = ['thumbs_up', 'peace', 'stop', 'fist', 'palm']

# Mapping vers contexte médical
gesture_mapping = {
    'thumbs_up': {
        'medical_label': 'confident',
        'description': 'Confiance, approbation, attitude positive',
        'impact_score': +0.5
    },
    'peace': {
        'medical_label': 'open',
        'description': 'Ouverture, décontraction, communication fluide',
        'impact_score': +0.3
    },
    'palm': {
        'medical_label': 'explaining',
        'description': 'Explication, présentation, geste pédagogique',
        'impact_score': +0.3
    },
    'stop': {
        'medical_label': 'defensive',
        'description': 'Attitude défensive, refus, fermeture',
        'impact_score': -0.4
    },
    'fist': {
        'medical_label': 'stressed',
        'description': 'Stress, tension, inconfort',
        'impact_score': -0.5
    }
}

print("Gesture Categories Loaded:")
for cat in gesture_categories:
    print(f"  • {cat} → {gesture_mapping[cat]['medical_label']} ({gesture_mapping[cat]['impact_score']:+.1f})")
````

---
## 2. Explore Dataset Structure

````python
# Analyser la structure du dataset
dataset_info = []

for category in gesture_categories:
    cat_path = dataset_path / category
    
    if cat_path.exists():
        # Compter les images
        images = list(cat_path.glob('*.jpg')) + list(cat_path.glob('*.png'))
        num_images = len(images)
        
        # Info sur la première image
        if images:
            sample_img = Image.open(images[0])
            width, height = sample_img.size
            format_img = sample_img.format
        else:
            width, height, format_img = None, None, None
        
        dataset_info.append({
            'category': category,
            'medical_label': gesture_mapping[category]['medical_label'],
            'num_images': num_images,
            'sample_width': width,
            'sample_height': height,
            'format': format_img
        })
    else:
        print(f"⚠️ Category folder not found: {category}")

# Créer DataFrame
df_info = pd.DataFrame(dataset_info)

print("\nDataset Structure:")
print(df_info)
print(f"\nTotal Images: {df_info['num_images'].sum()}")
print(f"Total Categories: {len(df_info)}")
````

````python
# Visualiser la distribution
plt.figure(figsize=(12, 6))
colors = ['green', 'blue', 'orange', 'red', 'purple']
plt.bar(df_info['category'], df_info['num_images'], color=colors)
plt.title('Distribution of Images per Gesture Category', fontsize=14, fontweight='bold')
plt.xlabel('Gesture Category', fontsize=12)
plt.ylabel('Number of Images', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Ajouter les nombres sur les barres
for i, v in enumerate(df_info['num_images']):
    plt.text(i, v + 10, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
````

---
## 3. Visualize Sample Images

````python
# Afficher des exemples d'images pour chaque catégorie
fig, axes = plt.subplots(len(gesture_categories), 5, figsize=(15, 12))
fig.suptitle('Sample Images per Gesture Category', fontsize=16, fontweight='bold')

for i, category in enumerate(gesture_categories):
    cat_path = dataset_path / category
    images = list(cat_path.glob('*.jpg'))[:5]  # Prendre 5 premières images
    
    for j, img_path in enumerate(images):
        img = Image.open(img_path)
        axes[i, j].imshow(img)
        axes[i, j].axis('off')
        
        if j == 0:
            axes[i, j].set_title(f"{category}\n({gesture_mapping[category]['medical_label']})", 
                                 fontweight='bold', fontsize=10)

plt.tight_layout()
plt.show()
````

---
## 4. Analyze Image Properties

````python
# Analyser les dimensions des images
image_dimensions = []

for category in gesture_categories:
    cat_path = dataset_path / category
    images = list(cat_path.glob('*.jpg'))[:100]  # Échantillon de 100 images
    
    for img_path in images:
        img = Image.open(img_path)
        width, height = img.size
        image_dimensions.append({
            'category': category,
            'width': width,
            'height': height,
            'aspect_ratio': width / height
        })

df_dims = pd.DataFrame(image_dimensions)

print("Image Dimensions Analysis (sample of 100 per category):")
print(df_dims.groupby('category')[['width', 'height']].describe())
````

````python
# Vérifier la cohérence des dimensions
unique_dims = df_dims.groupby('category')[['width', 'height']].nunique()

print("\nDimension Consistency Check:")
print(unique_dims)

if (unique_dims == 1).all().all():
    print("\n✓ All images have consistent dimensions within each category!")
else:
    print("\n⚠️ Images have varying dimensions - preprocessing needed")
````

````python
# Visualiser la distribution des dimensions
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Distribution des largeurs
for category in gesture_categories:
    cat_data = df_dims[df_dims['category'] == category]
    ax1.hist(cat_data['width'], alpha=0.5, label=category, bins=20)
ax1.set_title('Image Width Distribution', fontweight='bold')
ax1.set_xlabel('Width (pixels)')
ax1.set_ylabel('Frequency')
ax1.legend()

# Distribution des hauteurs
for category in gesture_categories:
    cat_data = df_dims[df_dims['category'] == category]
    ax2.hist(cat_data['height'], alpha=0.5, label=category, bins=20)
ax2.set_title('Image Height Distribution', fontweight='bold')
ax2.set_xlabel('Height (pixels)')
ax2.set_ylabel('Frequency')
ax2.legend()

plt.tight_layout()
plt.show()
````

---
## 5. Class Balance Analysis

````python
# Analyser l'équilibre des classes
print("Class Balance Analysis:")
print("="*60)

min_images = df_info['num_images'].min()
max_images = df_info['num_images'].max()
balance_ratio = max_images / min_images if min_images > 0 else float('inf')

print(f"Smallest class: {df_info.loc[df_info['num_images'].idxmin(), 'category']} with {min_images} images")
print(f"Largest class: {df_info.loc[df_info['num_images'].idxmax(), 'category']} with {max_images} images")
print(f"Balance ratio: {balance_ratio:.2f}")

if balance_ratio < 1.2:
    print("\n✓ Excellent balance (ratio < 1.2)")
    print("→ No special handling needed")
elif balance_ratio < 1.5:
    print("\n✓ Good balance (ratio < 1.5)")
    print("→ Minor adjustments may help")
elif balance_ratio < 2.0:
    print("\n⚠️ Moderate imbalance (ratio < 2.0)")
    print("→ Consider data augmentation for minority classes")
else:
    print("\n❌ Significant imbalance (ratio ≥ 2.0)")
    print("→ Data augmentation or class weighting required")
````

````python
# Visualiser l'équilibre avec pourcentages
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Pie chart
colors_pie = ['green', 'blue', 'orange', 'red', 'purple']
ax1.pie(df_info['num_images'], labels=df_info['category'], autopct='%1.1f%%', 
        colors=colors_pie, startangle=90)
ax1.set_title('Class Distribution (Percentage)', fontweight='bold')

# Bar chart with percentage
total = df_info['num_images'].sum()
percentages = (df_info['num_images'] / total * 100)
ax2.barh(df_info['category'], percentages, color=colors_pie)
ax2.axvline(x=20, color='red', linestyle='--', label='Perfect balance (20%)')
ax2.set_xlabel('Percentage (%)')
ax2.set_title('Class Balance Check', fontweight='bold')
ax2.legend()
ax2.invert_yaxis()

plt.tight_layout()
plt.show()
````

````python
# Créer un mapping complet avec statistiques
medical_mapping = []

for _, row in df_info.iterrows():
    category = row['category']
    medical_mapping.append({
        'Kaggle_Gesture': category,
        'Medical_Label': gesture_mapping[category]['medical_label'],
        'Description': gesture_mapping[category]['description'],
        'Impact_Score': gesture_mapping[category]['impact_score'],
        'Num_Images': row['num_images'],
        'Use_Case': 'Training evaluation' if gesture_mapping[category]['impact_score'] != 0 else 'Neutral behavior'
    })

df_mapping = pd.DataFrame(medical_mapping)

print("\nGesture Mapping for Medical Representative Training:")
print("="*80)
print(df_mapping.to_string(index=False))
````

````python
# Visualiser l'impact des gestes
plt.figure(figsize=(12, 6))

# Trier par impact score
df_sorted = df_mapping.sort_values('Impact_Score')

colors_impact = ['red' if x < 0 else 'green' for x in df_sorted['Impact_Score']]
plt.barh(df_sorted['Medical_Label'], df_sorted['Impact_Score'], color=colors_impact)
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
plt.xlabel('Impact on Performance Score', fontsize=12)
plt.ylabel('Gesture (Medical Context)', fontsize=12)
plt.title('Impact of Detected Gestures on Overall Evaluation', fontsize=14, fontweight='bold')
plt.grid(axis='x', alpha=0.3)

# Annotations
for i, v in enumerate(df_sorted['Impact_Score']):
    plt.text(v, i, f' {v:+.1f}', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
````

## 7. Train/Val/Test Split Recommendation

````python
# Calculer les splits recommandés
total_images = df_info['num_images'].sum()

train_pct = 0.70
val_pct = 0.15
test_pct = 0.15

print("Recommended Dataset Split:")
print("="*60)
print(f"Total images: {total_images}")
print(f"\nTraining set ({train_pct*100:.0f}%): {int(total_images * train_pct)} images")
print(f"Validation set ({val_pct*100:.0f}%): {int(total_images * val_pct)} images")
print(f"Test set ({test_pct*100:.0f}%): {int(total_images * test_pct)} images")

print("\nPer-category split:")
print("-"*60)

split_details = []
for _, row in df_info.iterrows():
    n = row['num_images']
    split_details.append({
        'Category': row['category'],
        'Total': n,
        'Train': int(n * train_pct),
        'Val': int(n * val_pct),
        'Test': int(n * test_pct)
    })

df_split = pd.DataFrame(split_details)
print(df_split.to_string(index=False))
````

````python
# Visualiser le split
fig, ax = plt.subplots(figsize=(12, 6))

categories = df_split['Category']
train_counts = df_split['Train']
val_counts = df_split['Val']
test_counts = df_split['Test']

x = np.arange(len(categories))
width = 0.25

ax.bar(x - width, train_counts, width, label='Train (70%)', color='green')
ax.bar(x, val_counts, width, label='Validation (15%)', color='orange')
ax.bar(x + width, test_counts, width, label='Test (15%)', color='red')

ax.set_xlabel('Category')
ax.set_ylabel('Number of Images')
ax.set_title('Train/Val/Test Split per Category', fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
````

## 8. Data Augmentation Strategy

### 🔄 Recommended Augmentation Techniques:

Avec ~2000 images, l'augmentation est **IMPORTANTE** pour améliorer la robustesse du modèle.

**Augmentations Recommandées:**

1. **Rotation**: ±15° (mains peuvent être légèrement inclinées)
2. **Horizontal Flip**: Oui (main droite ↔ main gauche)
3. **Zoom**: 0.8-1.2 (distance variable de la caméra)
4. **Brightness**: ±20% (conditions d'éclairage variables)
5. **Shift**: ±10% horizontal/vertical (position dans le cadre)
6. **Shear**: ±5° (angle de vue variable)

**À ÉVITER:**
- ❌ Vertical flip (main ne se retourne pas)
- ❌ Rotation > 30° (geste devient méconnaissable)
- ❌ Distorsion excessive (dégrade la qualité)

**Objectif d'Augmentation:**
- Multiplier par 3-5x → ~6,000-10,000 images
- Appliquer UNIQUEMENT sur training set
- Garder validation/test sets originaux

````python
# Code exemple d'augmentation (à utiliser dans Data Preparation)
augmentation_code = '''
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Configuration d'augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,                    # Normalisation
    rotation_range=15,                 # Rotation ±15°
    width_shift_range=0.1,             # Shift horizontal 10%
    height_shift_range=0.1,            # Shift vertical 10%
    shear_range=0.1,                   # Shear 10%
    zoom_range=0.2,                    # Zoom 80-120%
    horizontal_flip=True,              # Flip horizontal
    brightness_range=[0.8, 1.2],       # Brightness ±20%
    fill_mode='nearest'                # Remplissage pixels
)

# Validation/Test: seulement normalisation
val_test_datagen = ImageDataGenerator(rescale=1./255)

# Charger les données
train_generator = train_datagen.flow_from_directory(
    'gesture_dataset_kaggle/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
'''

print("Example Augmentation Code for Data Preparation Phase:")
print("="*60)
print(augmentation_code)
````

## 9. Model Architecture Recommendations

### 🤖 Recommended CNN Architectures:

**Option 1: Transfer Learning (RECOMMANDÉ)**

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout

# Base model pré-entraîné
base_model = EfficientNetB0(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)

# Geler les couches de base
base_model.trainable = False

# Ajouter des couches personnalisées
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(5, activation='softmax')  # 5 classes
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

**Alternatives:**
- **ResNet50**: Robuste, bien testé
- **MobileNetV2**: Léger, rapide (pour production)
- **VGG16**: Simple, bon pour baseline

**Objectifs de Performance:**
- Accuracy validation: **>90%**
- Accuracy test: **>85%**
- F1-Score par classe: **>0.80**
- Temps d'inférence: **<100ms**

**Training Configuration:**
```python
# Callbacks
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

callbacks = [
    EarlyStopping(patience=10, restore_best_weights=True),
    ModelCheckpoint('best_model.h5', save_best_only=True),
    ReduceLROnPlateau(factor=0.5, patience=5)
]

# Training
history = model.fit(
    train_generator,
    epochs=50,
    validation_data=val_generator,
    callbacks=callbacks
)
```

```python
def evaluate_medical_rep(video_path, audio_path, transcript):
    # 1. Évaluation gestuelle (CNN)
    gesture_score = gesture_model.predict(video_frames)
    gesture_impact = map_gesture_to_impact(gesture_score)
    
    # 2. Évaluation verbale (NLP)
    text_score = nlp_model.predict(transcript)
    
    # 3. Évaluation vocale (ASR + prosody)
    voice_score = voice_model.predict(audio_features)
    
    # 4. Fusion des scores
    final_score = (
        text_score * 0.50 +      # 50% contenu verbal
        gesture_impact * 0.30 +  # 30% langage corporel
        voice_score * 0.20       # 20% ton/débit
    )
    
    # 5. Génération feedback
    feedback = generate_feedback(text_score, gesture_impact, voice_score)
    
    return {
        'final_score': final_score,
        'text_score': text_score,
        'gesture_impact': gesture_impact,
        'voice_score': voice_score,
        'feedback': feedback
    }
```
